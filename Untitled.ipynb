```python
import matplotlib.pyplot as plt
import networkx as nx

G=nx.Graph()
f = open('./edgelist.txt')
for line in f:
    A=list(map(str,line.split()))
    G.add_edge(A[0],A[1],weight=int(A[2]))
```

```python
nx.draw(G)
plt.show() #Считать и отобразить граф городов;
```

```python
g=nx.dfs_tree(G,'Ананасовый')
nx.draw(g)
plt.show() 
#Построить и отобразить остовное дерево методом обхода в ширину (DFS);
```

```python
h=nx.bfs_tree(G,'Ананасовый')
nx.draw(h)
plt.show() 
#Построить и отобразить остовное дерево методом обхода в глубину (BFS);
```

```python
print(nx.is_connected(G))
graphs = list(nx.connected_component_subgraphs(G))
for graph in graphs:
    nx.draw(graph)
    plt.show() 
#Выделить и отобразить компоненты связности, проверить связность графа;
```

```python
length,path = nx.single_source_dijkstra(G,'Ананасовый',cutoff=None)
print(length) 
#Найти длины кратчайших путей от заданной вершины ко всем остальным
```

```python
A=nx.dijkstra_path(G,'Ананасовый','Терновый')
print(A)
pos=nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, nodelist=G, node_size=100, node_color='r')
nx.draw_networkx_edges(G, pos, width=3.0, alpha=0.5)
for nodes in A:
    nx.draw_networkx_nodes(G,pos,nodelist=A,node_size=150,node_color='b')
plt.axis('off')
plt.show()
#Отобразить кратчайший путь между двумя вершинами (восстановление пути);
```
